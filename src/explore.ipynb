```python
data = {
    "Nombre de la Cancion": [],
    "Popularidad": [],
    "Duracion (minutos)": []
}


for index, track in enumerate(top_tracks['tracks'][:10], start=1):
    nombre_cancion = track['name']
    popularidad = track['popularity']
    duracion_ms = track['duration_ms']
    duracion_minutos = duracion_ms / 1000 / 60
    
    
    data["Nombre de la Cancion"].append(nombre_cancion)
    data["Popularidad"].append(popularidad)
    data["Duracion (minutos)"].append(duracion_minutos)


df = pd.DataFrame(data)


df_sorted = df.sort_values(by='Popularidad', ascending=False)

print(df_sorted.head(3))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.scatter(df['Duracion (minutos)'], df['Popularidad'])
plt.title('Radiohead')
plt.xlabel('Duracion (minutos)')
plt.ylabel('Popularidad')
plt.grid(True)
plt.show()
```
